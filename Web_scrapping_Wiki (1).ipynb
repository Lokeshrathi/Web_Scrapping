```python
#import the library to query a website
import requests
#specify the URL
wiki_link="https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
#get the content of the link
link=requests.get(wiki_link).content
```

```python
print(link)
```

```python
#Create a object named soup using BeautifulSoup libraray for extracting HTML or Xml files
from bs4 import BeautifulSoup
soup = BeautifulSoup(link, 'lxml')
print(soup)
```

```python
# to make it more readable with indentations
print(soup.prettify())
```

```python
# lets check if we did get the correct Link
soup.title
```

```python
# to check only the title
soup.title.string
```

```python
#finding all the <a> tags present in the page.
soup.find_all("a")
```

```python
# finding all the <a> tags that represents a links present on the page
all_link=soup.find_all("a")
for link in all_link:
    print(link.get("href"))
```

```python
# finding all the tables present in the page
all_tables=soup.find_all('table')
print(all_tables)
```

```python
#finding the table from where we like to extract the names of the countries
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
#finding the table links
table_links = right_table.findAll('a')
table_links
```

```python
#create a list country to store all the values from the table
country =[]
for links in table_links:
    country.append(links.get('title')) #since all the Countries name has 'title'
print(country)
```

```python
#representing the lists in the form of DataFrame
import pandas as pd
df = pd.DataFrame()
df['Country']=country
df
```

```python
#to remove all the NaN values lets make 'country' as the index
df.set_index('Country')
```

```python
#dropping the NaN values
df.dropna(inplace=True)
```

```python
#final output with the names of the country and no Null Values
df
```
